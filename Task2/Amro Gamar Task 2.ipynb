# Step 3: Feature Engineering

### Encode Categorical Variables
our dataset has categorical columns like:

- type of meal
- room type
- market segment type
- booking status



```python
encoded_data = pd.get_dummies(data, columns=['type of meal', 'room type', 'market segment type', 'booking status'], drop_first=True)
encoded_data
```

### Extract Date-Related Features
our dataset contains a column date of reservation, which we should convert to datetime and extract useful information like year, month, day, weekday, weekend/weekday.

```python
encoded_data['date of reservation'] = pd.to_datetime(encoded_data['date of reservation'], errors='coerce')

encoded_data['year'] = encoded_data['date of reservation'].dt.year
encoded_data['month'] = encoded_data['date of reservation'].dt.month
encoded_data['day_of_week'] = encoded_data['date of reservation'].dt.dayofweek  # 0 = Monday, 6 = Sunday
encoded_data['is_weekend'] = encoded_data['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)  # 1 if weekend, 0 if weekday

encoded_data = encoded_data.drop(columns=['date of reservation'])
```

### Createing a New Feature: 
Total guests = number of adults + number of children

```python
encoded_data['total_guests'] = encoded_data['number of adults'] + encoded_data['number of children']
encoded_data.columns
```
