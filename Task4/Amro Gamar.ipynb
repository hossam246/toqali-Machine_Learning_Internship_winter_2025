# Step 2: Data preperation

Convert pickup_datetime to datetime format

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])
```

key is just a unique identifier

```python
df.drop(columns=['key'], inplace=True)
```

Convert Car Condition, Weather, and Traffic Condition to categorical

```python
df['Car Condition'] = df['Car Condition'].astype('category')
df['Weather'] = df['Weather'].astype('category')
df['Traffic Condition'] = df['Traffic Condition'].astype('category')
```

```python
# rename 
df.rename(columns= {'distance' : 'trip distance'})
```

# Step 3: understanding fetaures

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
print(numerical_cols)
```

```python
sns.set_style("whitegrid")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.histplot(df['fare_amount'], bins=50, kde=True, ax=axes[0], color="blue")
axes[0].set_title("Distribution of Fare Amounts")
axes[0].set_xlabel("Fare Amount ($)")
axes[0].set_ylabel("Frequency")

sns.boxplot(x=df['fare_amount'], ax=axes[1], color="orange")
axes[1].set_title("Box Plot of Fare Amounts")
axes[1].set_xlabel("Fare Amount ($)")

plt.tight_layout()
plt.show()
```

```python
for col in numerical_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col], kde=True, bins=30, color="blue")
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
df.drop(columns=['User ID', 'Driver Name', 'User Name'], inplace=True)
```

```python
categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print(categorical_cols)
```

```python
for col in categorical_cols:
    print(f"{col}: {df[col].nunique()} unique values")
```

```python
for col in categorical_cols:
    print(f"\n{col} Distribution:")
    print(df[col].value_counts())
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

# Compute correlation again
plt.figure(figsize=(12, 6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix (With Label Encoded Features)")
plt.show()
```

```python
monthly_trends = df.groupby(['year', 'month'])['fare_amount'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_trends, x='month', y='fare_amount', hue='year', marker="o")
plt.title("Time Series Analysis â€“ Fare Amount Trends Over the Year")
plt.xlabel("Month")
plt.ylabel("Average Fare Amount ($)")
plt.legend(title="Year")
plt.grid(True)
plt.show()
```

```python

df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'])

df['hour'] = df['pickup_datetime'].dt.hour
df['weekday'] = df['pickup_datetime'].dt.weekday  # Monday = 0, Sunday = 6

df_weekdays = df[df['weekday'] < 5]
df_weekends = df[df['weekday'] >= 5]

fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

sns.countplot(x=df_weekdays['hour'], palette="Blues", ax=axes[0])
axes[0].set_title("Weekday Peak Hours")
axes[0].set_xlabel("Hour of the Day")
axes[0].set_ylabel("Number of Rides")
axes[0].set_xticks(range(0, 24))

sns.countplot(x=df_weekends['hour'], palette="Oranges", ax=axes[1])
axes[1].set_title("Weekend Peak Hours")
axes[1].set_xlabel("Hour of the Day")
axes[1].set_xticks(range(0, 24))

plt.suptitle("Comparing Peak Hours: Weekdays vs. Weekends", fontsize=14)
plt.show()
```

```python
# Plot ride counts by weather
plt.figure(figsize=(12, 5))
sns.countplot(x=df['Weather'], order=df['Weather'].value_counts().index, palette="coolwarm")
plt.xlabel("Weather Condition")
plt.ylabel("Number of Rides")
plt.title("Impact of Weather on Ride Demand")
plt.xticks(rotation=45)
plt.show()

# Plot ride counts by traffic condition
plt.figure(figsize=(12, 5))
sns.countplot(x=df['Traffic Condition'], order=df['Traffic Condition'].value_counts().index, palette="magma")
plt.xlabel("Traffic Condition")
plt.ylabel("Number of Rides")
plt.title("Impact of Traffic on Ride Demand")
plt.xticks(rotation=45)
plt.show()

# Plot ride counts by traffic condition
plt.figure(figsize=(12, 5))
sns.countplot(x=df['Car Condition'], order=df['Car Condition'].value_counts().index, palette="magma")
plt.xlabel("Traffic Condition")
plt.ylabel("Number of Rides")
plt.title("Impact of Car Condition on Ride Demand")
plt.xticks(rotation=45)
plt.show()
```
