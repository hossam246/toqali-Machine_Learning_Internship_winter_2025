```python
numerical_columns = [
       'pickup_longitude', 'pickup_latitude', 'dropoff_longitude',
       'dropoff_latitude', 'passenger_count', 'hour', 'day', 'month',
       'weekday', 'year', 'jfk_dist', 'ewr_dist', 'lga_dist', 'sol_dist',
       'nyc_dist', 'distance', 'bearing']
```

```python
categorical_columns = ["Car Condition", "Weather", "Traffic Condition"]
```

```python
target_column = 'fare_amount'
```

```python
print("Univariate Analysis: Categorical  Columns")
c = 0
for col in categorical_columns:
    c += 1
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.savefig(f'D:\\cellula\\{c}.png')
    plt.show()
```

```python
print("Univariate Analysis: Numerical Columns")
for col in numerical_columns:
    c += 1
    plt.figure(figsize=(8, 6))
    sns.histplot(df[col].dropna(), kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.savefig(f'D:\\cellula\\{c}.png')
    plt.show()
```

```python
print("Correlation Heatmap")
plt.figure(figsize=(12, 10))
sns.set(font_scale=1.2)
corr = df[numerical_columns].corr()
c += 1
# Plot the heatmap
heatmap = sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    fmt=".2f",
    annot_kws={"size": 12},
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)


plt.title('Correlation Heatmap', fontsize=16, pad=20)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig(f'D:\\cellula\\{c}.png')
plt.show()

```

```python
print("Outlier Detection")
for col in numerical_columns:
    c += 1
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col} (Outlier Detection)')
    plt.savefig(f'D:\\cellula\\{c}.png')
    plt.show()
```

```python
print("Bivariate Analysis: Categorical vs. Categorical")
for i in range(len(categorical_columns)):
    for j in range(i + 1, len(categorical_columns)):
        c += 1
        plt.figure(figsize=(8, 6))
        sns.heatmap(pd.crosstab(df[categorical_columns[i]], df[categorical_columns[j]]), annot=True, fmt='d', cmap='Blues')
        plt.title(f'{categorical_columns[i]} vs {categorical_columns[j]}')
        plt.savefig(f'D:\\cellula\\{c}.png')
        plt.show()
```
