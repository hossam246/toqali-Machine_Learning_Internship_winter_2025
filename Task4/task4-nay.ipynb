# Explaining columns name

 - User_ID: A unique number for each user.
 - User Name: The user's name.
 - Driver Name: The driver's name.
 - Car Condition: Whether the car is in good or bad condition.
 - Weather: The weather condition (winter, windy, sunny, etc.).
 - Traffic Conditions: Whether traffic is congested or normal.
 - Key: A unique number for each trip.
 - Fare Amount: The trip fare.
 - Pickup Datetime: The time when the passenger requested Uber.
 - Pickup Longitude & Latitude: The starting location (passenger’s location).
 - Drop-off Longitude & Latitude: The destination location (where the passenger was dropped off).
 - Passenger Count: The number of passengers in the car.
Hour, Day, Month, Week: Repeated time details for when the passenger requested Uber.
 - JFK, EWR, LGA Distance: The distance between the pickup location and these airports.
 - SOL Distance: The distance from the Statue of Liberty.
 - Distance: The trip distance.
 - Bearing: The trip direction from start to end, measured in angles.

```python
#Which airport route (JFK, EWR, LGA) has the highest and lowest fares?

# Calculate the average fare for each airport
airport_fares = {
    "JFK": data.loc[data["jfk_dist"] > 0, "fare_amount"].mean(),
    "EWR": data.loc[data["ewr_dist"] > 0, "fare_amount"].mean(),
    "LGA": data.loc[data["lga_dist"] > 0, "fare_amount"].mean()
}

# Convert to Series for plotting
airport_fares = pd.Series(airport_fares)

# Plot
plt.figure(figsize=(8, 5))
airport_fares.plot(kind="bar", color=["orchid", "lightblue", "pink"])

# Labels and title
plt.xlabel("Airport")
plt.ylabel("Average Fare Amount")
plt.title("Average Fare per Airport Route")

# Show plot
plt.show()
```

### **Fare Per Kilometer Analysis**  

- When analyzing `fare_per_km`, a **clear inverse relationship** was observed: **as distance increases, fare per km decreases**.  
- The likely reason is that **short-distance trips dominate the dataset**, causing the **overall average fare per km** to be skewed higher for them.  
- Possible contributing factors:  
  - **Base fare effect:** Short trips include a fixed base fare, making their per km rate appear higher.  
  - **Fewer long trips:** The number of rides significantly decreases as distance increases, leading to a lower per km fare for those trips.  
  - **Potential discounts:** Longer trips may include reduced rates per km, contributing to the decreasing trend.  

**Conclusion:** A bar chart of ride counts across distance ranges showed that **most trips are short**, supporting this explanation.

# Time-Based Analysis

```python
month_counts = data['month'].value_counts().sort_index()

plt.figure(figsize=(8, 4))
sns.barplot(x=month_counts.index, y=month_counts.values, color='orchid')
plt.xlabel('Month')
plt.ylabel('Number of Rides')
plt.title('Number of Rides per Month')
plt.show()
```

- The number of rides is **consistent across most months**, However, there is a **slight drop after June**.  


```python
data['week'] = pd.to_datetime(data['pickup_datetime']).dt.isocalendar().week

week_counts = data['week'].value_counts().sort_index()

plt.figure(figsize=(8, 4))
sns.lineplot(x=week_counts.index, y=week_counts.values, marker='o', color='orchid')
plt.xlabel('Week Number')
plt.ylabel('Number of Rides')
plt.title('Number of Rides per Week')
plt.show()
```

- The number of rides remains high throughout the year but **starts to drop after week 50**.  
- Possible reasons for the decline:  
  - **Holidays (Christmas & New Year’s Eve)** reduce daily commuting.  
  - **Winter weather** may discourage people from going out.  
  - **Shorter weeks (week 51 & 52)** might have fewer recorded days.  
  - **Changes in travel patterns** (school breaks, remote work) could impact ride demand.
