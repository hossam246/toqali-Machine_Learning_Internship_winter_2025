# Step 3: Feature Engineering

### Encode Categorical Variables
our dataset has categorical columns like:

- type of meal
- room type
- market segment type
- booking status



```python
encoded_data = pd.get_dummies(data, columns=['type of meal', 'room type', 'market segment type', 'booking status'], drop_first=True)
encoded_data
```

### Extract Date-Related Features
our dataset contains a column date of reservation, which we should convert to datetime and extract useful information like year, month, day, weekday, weekend/weekday.

```python
encoded_data['date of reservation'] = pd.to_datetime(encoded_data['date of reservation'], errors='coerce')

encoded_data['year'] = encoded_data['date of reservation'].dt.year
encoded_data['month'] = encoded_data['date of reservation'].dt.month
encoded_data['day_of_week'] = encoded_data['date of reservation'].dt.dayofweek  # 0 = Monday, 6 = Sunday
encoded_data['is_weekend'] = encoded_data['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)  # 1 if weekend, 0 if weekday

encoded_data = encoded_data.drop(columns=['date of reservation'])
```

### Createing a New Feature: 
Total guests = number of adults + number of children

```python
encoded_data['total_guests'] = encoded_data['number of adults'] + encoded_data['number of children']
encoded_data.columns
```

# Step 5: EDA Exploratory Data Analysis

### Outliers detection and handling

```python
num_cols = ['number of weekend nights',
            'number of week nights', 'lead time', 'average price ', 
            'special requests']

for col in num_cols:

    plt.figure(figsize=(15, 4))
    
    plt.subplot(1, 2, 1)
    data[col].hist(grid=False)
    plt.title(f'Histogram of {col}')
    plt.ylabel('Count')
    
    plt.subplot(1, 2, 2)
    sns.boxplot(x=encoded_data[col])
    plt.title(f'Boxplot of {col}')
    
    plt.show()
```

```python
plt.figure(figsize=(12, 6))

# Histogram
plt.figure(figsize=(12, 6))
sns.histplot(encoded_data['average price '], bins=50, kde=True)
plt.title('Histogram of Lead Time')
plt.show()

```

```python
encoded_data['lead time'] = np.log1p(encoded_data['lead time'])
encoded_data['average price '] = np.log1p(encoded_data['average price '])
```

```python
# Histogram of lead time after transformation
plt.figure(figsize=(12, 6))
sns.histplot(encoded_data['lead time'], bins=50, kde=True)
plt.title('Histogram of Transformed Lead Time')
plt.show()

# Histogram of average price after transformation
plt.figure(figsize=(12, 6))
sns.histplot(encoded_data['average price '], bins=50, kde=True)
plt.title('Histogram of Transformed Lead Time')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 10))
fig.suptitle('Bar plot for all categorical variables in the dataset')

sns.countplot(ax=axes[0, 0], x='type of meal', data=data, color='blue', 
              order=data['type of meal'].value_counts().index)
sns.countplot(ax=axes[0, 1], x='room type', data=data, color='blue', 
              order=data['room type'].value_counts().index)
sns.countplot(ax=axes[1, 0], x='market segment type', data=data, color='blue', 
              order=data['market segment type'].value_counts().index)
sns.countplot(ax=axes[1, 1], x='booking status', data=data, color='blue', 
              order=data['booking status'].value_counts().index)

axes[1][0].tick_params(labelrotation=45)
axes[1][1].tick_params(labelrotation=45)

plt.tight_layout()
plt.show()
```

## Observations and insights:

### Type of Meal

- "Not Selected" the most common category.
- "Meal Plan 1" is the second most popular .

### Room Type

- "Room_Type 1" the most booked room.
- "Room_Type 2" second most booked, while "Room_Type 7" are rare.

### Market Segment Type

- "Online" most reservations are made through online channels.
- "Offline" a significant channel, but smaller than online.
- "Corporate" and "Complementary" have fewer bookings.


```python
fig, axarr = plt.subplots(4, 2, figsize=(12, 18))

data.groupby('room type')['average price '].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12)
axarr[0][0].set_title("Room Type Vs Price", fontsize=18)

data.groupby('market segment type')['average price '].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12)
axarr[0][1].set_title("Market Segment Type Vs Price", fontsize=18)

data.groupby('type of meal')['average price '].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12)
axarr[1][0].set_title("Meal Type Vs Price", fontsize=18)

data.groupby('booking status')['average price '].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12)
axarr[1][1].set_title("Booking Status Vs Price", fontsize=18)

data.groupby('room type')['average price '].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][0], fontsize=12)
axarr[2][0].set_title("Top Room Types Vs Price", fontsize=18)

data.groupby('lead time')['average price '].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][1], fontsize=12)
axarr[2][1].set_title("Lead Time Vs Price", fontsize=18)

data.groupby('number of adults')['average price '].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=12)
axarr[3][0].set_title("Number of Adults Vs Price", fontsize=18)

data.groupby('total_guests')['average price '].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=12)
axarr[3][1].set_title("Total Guests Vs Price", fontsize=18)

plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=0.5)

sns.despine()
```
