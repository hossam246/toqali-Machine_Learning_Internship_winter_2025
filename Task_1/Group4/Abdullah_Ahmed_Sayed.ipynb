```python
plt.figure(figsize=(6,4))
sns.countplot(x="booking status", data = data, palette="Set1")
plt.title("Booking Status Distribution")
plt.xlabel("Booking Status")
plt.ylabel("Count")
plt.show()
```

```python
sns.boxplot(data=data['average price '], palette="viridis")
```

```python
sns.boxplot(data=data['lead time'], palette="viridis")
```

```python
leadtime_median = data['lead time'].median()
averageprice_median = data['average price '].median()
leadtime_upper_bound = data['lead time'].quantile(0.75) + 1.5 * (data['lead time'].quantile(0.75) - data['lead time'].quantile(0.25))
leadtime_lower_bound = data['lead time'].quantile(0.25) - 1.5 * (data['lead time'].quantile(0.75) - data['lead time'].quantile(0.25))
averageprice_upper_bound = data['average price '].quantile(0.75) + 1.5 * (data['average price '].quantile(0.75) - data['average price '].quantile(0.25))
averageprice_lower_bound = data['average price '].quantile(0.25) - 1.5 * (data['average price '].quantile(0.75) - data['average price '].quantile(0.25))
```

```python
data['average price '] = np.where((data['average price '] > averageprice_upper_bound) | (data['average price '] < averageprice_lower_bound), averageprice_median, data['average price '])
data['lead time'] = np.where((data['lead time'] > leadtime_upper_bound) | (data['lead time'] < leadtime_lower_bound), leadtime_median, data['lead time'])
```

```python
sns.boxplot(data=data['average price '], palette="viridis")
```

```python
sns.boxplot(data=data['lead time'], palette="viridis")
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data["lead time"], bins=30, kde=False, color="red")
plt.title("Lead Time Distribution")
plt.xlabel("Lead Time (Days)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(x=data["type of meal"], y=data["average price "], palette="tab10", color="blue")
plt.title("Meal Plans vs. Average Price")
plt.xlabel("Meal Plans")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.lineplot(x=data["room type"], y=data["average price "], palette="tab10", color="black")
plt.title("Room Types vs. Average Price")
plt.xlabel("Room Types")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x=data['room type'], y=data["number of adults"], hue=data['number of children'], palette="tab10")
plt.title("Room Type vs. Number of Adults")
plt.xlabel("Room Type")
plt.ylabel("Number of Adults")
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.barplot(x=data['room type'], y=data["number of children"], hue=data['number of adults'], palette="Set3")
plt.title("Room Type vs. Number of Children")
plt.xlabel("Room Type")
plt.ylabel("Number of Children")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(data.select_dtypes(include=["int64", "float64"]).corr(), annot=True, cmap="rainbow", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
